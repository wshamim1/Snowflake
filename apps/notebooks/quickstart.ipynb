# Snowflake Quickstart Notebook
Use this notebook to test connection and query data from Snowflake.

```python
import os
import snowflake.connector
from dotenv import load_dotenv

load_dotenv()

# Connection that works without importing local src modules
conn = snowflake.connector.connect(
    user=os.getenv("SNOWFLAKE_USER"),
    password=os.getenv("SNOWFLAKE_PASSWORD"),
    account=os.getenv("SNOWFLAKE_ACCOUNT"),
    warehouse=os.getenv("SNOWFLAKE_WAREHOUSE"),
    database=os.getenv("SNOWFLAKE_DATABASE"),
    schema=os.getenv("SNOWFLAKE_SCHEMA"),
)

TABLE_NAME = os.getenv("SNOWFLAKE_TABLE", "TABLE1")


def fetch_data(query: str):
    try:
        cursor = conn.cursor()
        cursor.execute(query)
        return cursor.fetchall()
    except Exception as e:
        print(f"Query error: {e}")
        return None
    finally:
        try:
            cursor.close()
        except Exception:
            pass


def execute_dml(query: str) -> bool:
    try:
        cursor = conn.cursor()
        cursor.execute(query)
        conn.commit()
        return True
    except Exception as e:
        print(f"DML error: {e}")
        return False
    finally:
        try:
            cursor.close()
        except Exception:
            pass


print("Connected")
```

```python
result = fetch_data('SELECT CURRENT_VERSION();')
print(result)
```

## SELECT Operations
Query data from Snowflake tables.

```python
# Query all records from table
query = f"SELECT * FROM {TABLE_NAME} LIMIT 10;"
result = fetch_data(query)

if result:
    print(f"Found {len(result)} rows:")
    for row in result:
        print(row)
else:
    print("No data found or query failed")
```

## INSERT Operations
Add new records to Snowflake tables.

```python
# Insert a new record
insert_query = f"""
INSERT INTO {TABLE_NAME} (ID, NAME, CREATED_AT)
VALUES (999, 'Test from Notebook', CURRENT_TIMESTAMP());
"""

success = execute_dml(insert_query)

if success:
    print("✅ Insert successful!")
else:
    print("❌ Insert failed")
```

## UPDATE Operations
Modify existing records in Snowflake tables.

```python
# Update a record
update_query = f"""
UPDATE {TABLE_NAME}
SET NAME = 'Updated from Notebook'
WHERE ID = 999;
"""

success = execute_dml(update_query)

if success:
    print("✅ Update successful!")
else:
    print("❌ Update failed")
```

## DELETE Operations
Remove records from Snowflake tables.

```python
# Delete a record
delete_query = f"""
DELETE FROM {TABLE_NAME}
WHERE ID = 999;
"""

success = execute_dml(delete_query)

if success:
    print("✅ Delete successful!")
else:
    print("❌ Delete failed")
```

## Verify Changes
Query the table again to see the results of your operations.

```python
# Query all records to verify changes
verify_query = f"SELECT * FROM {TABLE_NAME} ORDER BY CREATED_AT DESC LIMIT 10;"
result = fetch_data(verify_query)

if result:
    print(f"Current records in {TABLE_NAME} (showing latest {len(result)}):\n")
    for idx, row in enumerate(result, 1):
        print(f"{idx}. {row}")
else:
    print("No data found")
```
