# Snowflake Standalone Notebook
This notebook is self-contained and can run directly in Snowflake Notebooks without external dependencies.

```python
import os
import snowflake.connector
from snowflake.snowpark import Session

# For Snowflake Notebooks, connection is automatic
# For local execution, use environment variables
try:
    # Try Snowflake Notebooks built-in session
    session = Session.builder.getOrCreate()
    print("Using Snowflake Notebooks session")
except:
    # Fallback to traditional connection for local execution
    from dotenv import load_dotenv
    load_dotenv()
    
    conn = snowflake.connector.connect(
        user=os.getenv("SNOWFLAKE_USER"),
        password=os.getenv("SNOWFLAKE_PASSWORD"),
        account=os.getenv("SNOWFLAKE_ACCOUNT"),
        warehouse=os.getenv("SNOWFLAKE_WAREHOUSE"),
        database=os.getenv("SNOWFLAKE_DATABASE"),
        schema=os.getenv("SNOWFLAKE_SCHEMA")
    )
    print("Using local connection")
```

## SELECT Query

```python
def run_query(query):
    """Execute SELECT query and return results"""
    try:
        cursor = conn.cursor()
        cursor.execute(query)
        results = cursor.fetchall()
        cursor.close()
        return results
    except Exception as e:
        print(f"Error: {e}")
        return None

# Test query
result = run_query("SELECT CURRENT_VERSION();")
print(result)
```

```python
# Query TABLE1
result = run_query("SELECT * FROM TABLE1 LIMIT 10;")
if result:
    for row in result:
        print(row)
```

## INSERT, UPDATE, DELETE

```python
def execute_dml(query):
    """Execute INSERT/UPDATE/DELETE queries"""
    try:
        cursor = conn.cursor()
        cursor.execute(query)
        conn.commit()
        cursor.close()
        return True
    except Exception as e:
        print(f"Error: {e}")
        return False

# Example INSERT
insert_query = "INSERT INTO TABLE1 (ID, NAME) VALUES (999, 'Snowflake Notebook');"
if execute_dml(insert_query):
    print("✅ Insert successful")

# Example UPDATE
update_query = "UPDATE TABLE1 SET NAME = 'Updated' WHERE ID = 999;"
if execute_dml(update_query):
    print("✅ Update successful")

# Example DELETE
delete_query = "DELETE FROM TABLE1 WHERE ID = 999;"
if execute_dml(delete_query):
    print("✅ Delete successful")
```
